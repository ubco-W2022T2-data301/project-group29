### On cleaning up and isolating the data needed 

```python
dataload = dataload_1[["Age", "Primary streaming service", "Hours per day", "Fav genre","BPM", "Exploratory", "Foreign languages"]]
dataload.dropna()
dataload
```

# Research Question

## How diferent are Gen Z's music habits from their Parents?

Ever since music has become more accessible, different age ranges listen to diffrent types of music. I wannna single out Gen Z as they have had the most access to music as compared to every other generation and compare tham to their Parents, who  had other not very accessible methods of listening to music. I want to compare the genres, how long they listen, the overall BPM of the music and their respective streaming services.

# Analysing the Data

```python
#Gen Z music 
age_z = dataload[["Age","Fav genre"]]
age_z = age_z[(age_z["Age"] <= 25)]



sns.boxplot(data = age_z, x = "Age", y = "Fav genre").set(title = "What Genres does Gen Z listen to", xlabel = "Age", ylabel = "Fav Genre")
plt.xticks(fontsize=12) and plt.yticks(fontsize=12) 
plt.show()
```

```python
GenZ_Genre = age_z["Fav genre"].mode()
GenZ_Genre
```

#### Most of Gen Z listens to almost all genres and is very diverse but the most Frequent out of them all is Rock

```python
#Gen X music 
age_x = dataload[["Age","Fav genre"]]
age_x = age_x[(age_x["Age"] >= 43) & (age_x["Age"] <= 60)]


sns.boxplot(data = age_x, x = "Age", y = "Fav genre").set(title = "What Genres does Gen X listen to", xlabel = "Age", ylabel = "Fav Genre")
plt.xticks(fontsize=12) and plt.yticks(fontsize=12)

plt.show()
```

```python
Genx_Genre = age_x["Fav genre"].mode()
Genx_Genre
```

#### Most of Gen X listens to Rock, Pop and R&B for the most part and its not very diverse. But the genre that is most frequent for them is Rock as well

```python
#Gen z hours
hour_z = dataload[["Age","Hours per day"]]
hour_z = hour_z[(hour_z["Age"] <=25)]

sns.regplot(data = hour_z, y = "Age", x = "Hours per day").set(title = "Hours Gen Z spends listening to music", ylabel = "Age", xlabel = "Hours per day")
plt.xticks(fontsize=12) and plt.yticks(fontsize=12)
plt.show()
```

```python
Genz_hours = hour_z["Hours per day"].mean()
Genz_hours
```

#### Gen Z on an average listens to about 3.7 hours of music in a day 

```python


#Gen x hours
hour_x = dataload[["Age","Hours per day"]]
hour_x = hour_x[(hour_x["Age"] <=60) & (hour_x["Age"] >= 43)]
#hour_x = hour_x[(hour_x["Age"] >= 43)]

sns.regplot(data = hour_x, y = "Age", x = "Hours per day").set(title = "Hours Gen X spends listening to music", ylabel = "Age", xlabel = "Hours per day")
plt.xticks(fontsize=12) and plt.yticks(fontsize=12)
plt.show()
```

#### Gen X on an average listens to 3 hours of music in a day.

```python
#Gen z Streraming service
st_z = dataload[["Age","Primary streaming service"]] 
st_z = st_z[(st_z["Age"] <= 25)]

plot_z = sns.jointplot(data = st_z, y = "Primary streaming service", x = "Age")
plot_z.fig.suptitle("Streaming service used by Gen Z", fontsize = 12)
plt.xticks(fontsize=12) and plt.yticks(fontsize=12)
plt.show()

```
