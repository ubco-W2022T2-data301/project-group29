# Research Question

## How different are Gen Z's music habits from their Parents?

Ever since music has become more accessible, different age ranges listen to diffrent types of music. I wannna single out Gen Z as they have had the most access to music as compared to every other generation and compare tham to their Parents, who  had other not very accessible methods of listening to music. I want to compare the genres, how long they listen, the overall BPM of the music and their respective streaming services.

# Analysing the Data

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,10))

age_z = dataload[["Age","Fav genre"]]
age_z = age_z[(age_z["Age"] <= 25)]

sns.boxplot(data = age_z, x = "Age", y = "Fav genre", ax=ax1).set(title = "What Genres does Gen Z listen to", xlabel = "Age", ylabel = "Fav Genre")

#Gen X music 
age_x = dataload[["Age","Fav genre"]]
age_x = age_x[(age_x["Age"] >= 43) & (age_x["Age"] <= 60)]

sns.boxplot(data = age_x, x = "Age", y = "Fav genre", ax=ax2).set(title = "What Genres does Gen X listen to", xlabel = "Age", ylabel = "Fav Genre")

plt.suptitle("Genre Divide between Gen Z and Gen X")
plt.savefig('../images/analysis2/genre_a2.png')

plt.tight_layout()
plt.show()
```

#### Most of Gen Z listens to almost all genres and is very diverse but the most Frequent out of them all is Rock. Most of Gen X listens to Rock, Pop and R&B for the most part and its not very diverse. But the genre that is most frequent for them is Rock as well

```python
GenZ_Genre = age_z["Fav genre"].mode()
GenZ_Genre
```

```python
Genx_Genre = age_x["Fav genre"].mode()
Genx_Genre
```

```python
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,10))

# Gen Z hours
hour_z = dataload[["Age","Hours per day"]]
hour_z = hour_z[(hour_z["Age"] <=25)]

# Plot the first regplot on the first subplot
sns.regplot(data = hour_z, y = "Age", x = "Hours per day", ax=ax1).set(title = "Hours Gen Z spends listening to music", ylabel = "Age", xlabel = "Hours per day")
ax1.tick_params(labelsize=12)

# Gen X hours
hour_x = dataload[["Age","Hours per day"]]
hour_x = hour_x[(hour_x["Age"] <=60) & (hour_x["Age"] >= 43)]

# Plot the second regplot on the second subplot
sns.regplot(data = hour_x, y = "Age", x = "Hours per day", ax=ax2).set(title = "Hours Gen X spends listening to music", ylabel = "Age", xlabel = "Hours per day")
ax2.tick_params(labelsize=12)

# Show the figure
plt.savefig('../images/analysis2/hours_a2.png')
plt.suptitle("Hours Spent Divide between Gen Z and Gen X")
plt.show()
```

#### Gen Z on an average listens to about 3.7 hours of music in a day. Gen X on an average listens to 3 hours of music in a day.

```python
Genz_hours = hour_z["Hours per day"].mean()
Genz_hours
```

```python
Genx_hours = hour_x["Hours per day"].mean()
Genx_hours
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,8))

# Gen z Streraming service
st_z = dataload[["Age","Primary streaming service"]] 
st_z = st_z[(st_z["Age"] <= 25)]
sns.scatterplot(data=st_z, x="Age", y="Primary streaming service", ax=ax1)
ax1.set(title="Streaming service used by Gen Z", xlabel="Age", ylabel="Primary streaming service")
ax1.tick_params(labelsize=12)

# Gen X streaming service
st_x = dataload[["Age","Primary streaming service"]]
st_x = st_x[(st_x["Age"] >= 43) & (st_x["Age"] <= 60)]
sns.scatterplot(data=st_x, x="Age", y="Primary streaming service", ax=ax2)
ax2.set(title="Streaming service used by Gen X", xlabel="Age", ylabel="Primary streaming service")
ax2.tick_params(labelsize=12)

plt.suptitle("Streaming Service Divide between Gen Z and Gen X")
plt.savefig('../images/analysis2/streaming_a2.png')
plt.tight_layout()
plt.show()
```

#### The most popular streaming service amongst gen Z is "Spotify". The most popular streaming service amongst gen X is "I do not use a streaming service"

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 10))
genz_bpm = dataload[["Age","BPM"]]
genz_bpm = genz_bpm[(genz_bpm["Age"] < 25) & (genz_bpm["BPM"] < 200)]

plot_bpmz = sns.barplot(data=genz_bpm , x = "Age", y = "BPM", ax = ax1)
ax1.set_title("Gen Z")

plot_bpmz

genx_bpm = dataload[["Age","BPM"]]
genx_bpm = genx_bpm[(genx_bpm["Age"] >= 43) & (genx_bpm["Age"] <= 60) & (genx_bpm["BPM"] < 200)]

plot_bpmx = sns.barplot(data=genx_bpm , x = "Age", y = "BPM", ax = ax2)
ax2.set_title("Gen X")
plot_bpmx

plt.suptitle("BPM Divide between Gen Z and Gen X")

plt.tight_layout()
plt.savefig('../images/analysis2/BPM_a2.png')
plt.show()
```

##### Gen Z usually listens to a lot of upbeat music ranging around 120 BPM. Gen X on the other hand listens to music that is not very upbeat around 95 BPM
