# Research Question

## How diiferent are Gen Z's music habits from their Parents?

Ever since music has become more accessible, different age ranges listen to diffrent types of music. I wannna single out Gen Z as they have had the most access to music as compared to every other generation and compare tham to their Parents, who  had other not very accessible methods of listening to music. I want to compare the genres, how long they listen, the overall BPM of the music and their respective streaming services.

```python
#Gen Z music 
age_z = dataload[["Age","Fav genre"]]
age_z = age_z[(age_z["Age"] <= 25)]

sns.boxplot(data = age_z, x = "Age", y = "Fav genre").plot
```

commentary: most of Gen Z listens to almost all genres and is very diverse.

```python
#Gen X music 
age_x = dataload[["Age","Fav genre"]]
age_x = age_x[(age_x["Age"] >= 43)]
age_x = age_x[(age_x["Age"] <= 60)]

sns.boxplot(data = age_x, x = "Age", y = "Fav genre").plot
```

Commentary: Most of Gen X listens to Rock, Pop and R&B for the most part. Not very diverse

```python
#Gen z hours
hour_z = dataload[["Age","Hours per day"]]
hour_z = hour_z[(hour_z["Age"] <=25)]

sns.lineplot(data = hour_z, x = "Age", y = "Hours per day").plot

```

Majority of Gen Z listens to about 5 hours a day with hours ranging from 0-10 in a day

```python


#Gen x hours
hour_x = dataload[["Age","Hours per day"]]
hour_x = hour_x[(hour_x["Age"] <=60)]
hour_x = hour_x[(hour_x["Age"] >= 43)]

sns.lineplot(data = hour_x, x = "Age", y = "Hours per day").plot

```

Majority of Gen x listens to about 2 hours a day with hours ranging from 0-10 in a day

```python
#Gen z Streraming service
st_z = dataload[["Age","Primary streaming service"]]
st_z = st_z[(st_z["Age"] <= 25)]

sns.jointplot(data = st_z, y = "Primary streaming service", x = "Age").plot

```

The most popular streaming service amongst gen Z is "Spotify"

```python
#Gen X streaming service
st_x = dataload[["Age","Primary streaming service"]]
st_x = st_x[(st_x["Age"] >= 43)]
st_x = st_x[(st_x["Age"] <= 60)]


sns.jointplot(data = st_x, y = "Primary streaming service", x = "Age" ).plot

```

The most popular streaming service amongst gen X is "Spotify" or "Other Streaming Service"
